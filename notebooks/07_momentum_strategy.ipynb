```python
df = yf.download(sp_2010_tickers, start=start_date)['Close']
```

```python
monthly_prices = df.resample('ME').last()
monthly_prices
```

```python
momentum = monthly_prices.pct_change(12).shift(1)
momentum = momentum.iloc[13:] # drop the first 12 months of NaN values
momentum = momentum.dropna(axis=1)
momentum # returns over the last 12 months for each month
```

```python
portfolio = pd.DataFrame(index=monthly_prices.index, columns=monthly_prices.columns)
portfolio
```

```python
constituents['date'] = pd.to_datetime(constituents['date'])
constituents = constituents.set_index('date')
```

```python
def get_tickers_date(df, date):
    if date >= df.index[-1]:
        ticks = df.iloc[-1].tickers
    else:
        ticks = df.loc[date:].iloc[0].tickers
    return ticks.split(',')

for date in momentum.index:
    pf_date = momentum.loc[date]
    relevant_tickets = get_tickers_date(constituents, date)
    pf_date = pf_date[pf_date.index.isin(relevant_tickets)]
    top20 = pf_date.nlargest(20)
    portfolio.loc[date, top20.index] = 1 / 20
```
