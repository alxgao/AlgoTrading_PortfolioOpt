# Engle–Granger test
$$a_t - \beta b_t = u_t$$
where $a$ and $b$ are distinct stocks, we regress to find $\beta$ and if $u_t$, the residuals, are stationary,

```python
actual_tickers = list(df.columns)
actual_tickers[:5]
```

```python
from sklearn.linear_model import LinearRegression
def find_beta(list_a, list_b):
    ln = LinearRegression()
    ln.fit(list_b, list_a)
    beta = ln.coef_[0][0]
    return beta
```

```python
# sample with first two stocks
a = df[actual_tickers[0]].values
b = df[actual_tickers[1]].values
beta = find_beta(a.reshape(-1, 1), b.reshape(-1, 1))
beta
```

```python
u = a - beta * b
u
```

### Augmented Dickey-Fuller unit root test

```python
from statsmodels.tsa.stattools import adfuller
adf, p_value, _, _, _, _ = adfuller(u)
adf, p_value
```

```python
# algorithm: most cointegrated are the stock pairs with the lowest adf among pairs with a statistically significant (alpha=0.05) p value
def engle_granger(list_a, list_b):
    beta = find_beta(list_a.reshape(-1, 1), list_b.reshape(-1, 1))
    u = list_a - beta * list_b
    adf, p_value, _, _, _, _ = adfuller(u)
    return adf, p_value

from itertools import combinations

sig_pairs = []
sig_adfs = []

for ticker_a, ticker_b in combinations(actual_tickers, 2):
    a = df[ticker_a].values
    b = df[ticker_b].values

    adf, p_value = engle_granger(a, b)
    if p_value <= 0.05: # statistically significant
        sig_pairs.append((ticker_a, ticker_b))
        sig_adfs.append(adf)
```

```python
adf_df = pd.DataFrame({"pair": sig_pairs, "adf": sig_adfs})
adf_df
```

```python
# getting the most cointegrated pair
adf_df_sorted = adf_df.sort_values("adf")
adf_df_sorted
```

```python
ticker_a, ticker_b = adf_df_sorted.iloc[0].pair
ticker_a, ticker_b
```

```python
beta = find_beta(df[ticker_a].values.reshape(-1 ,1), df[ticker_b].values.reshape(-1 ,1))
beta
```

```python
intercept = np.mean(df[ticker_a].values - beta * df[ticker_b].values)
intercept
```

```python
import backtrader as bt

class StatArb(bt.Strategy):
    rolling_window = 50    # days
    entry = 2               # z score
    exit = 0.5              # z score
    stop_loss = 4.0         # stop‑loss threshold (z)


    # simply prints
    def log(self, txt, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        print('%s, %s' % (dt.isoformat(), txt))

    # define parameters
    def __init__(self):
        self.order = None
        self.allow_entry = True

        self.a_close = self.datas[0].close
        self.b_close = self.datas[1].close

        self.spread = self.a_close - self.b_close * float(beta) - intercept

        self.spread_mean = bt.indicators.SMA(
            self.spread,
            period=self.rolling_window,
            plotname="SpreadMean"
        )

        self.spread_std = bt.indicators.StdDev(
            self.spread,
            period=self.rolling_window,
            plotname="SpreadStd"
        )

        self.zscore = (self.spread - self.spread_mean) / self.spread_std

    # tracking when the order is executed
    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return

        if order.status in [order.Completed]:
            if order.isbuy():
                self.log('BUY EXECUTED, %.2f' % order.executed.price)
            elif order.issell():
                self.log('SELL EXECUTED, %.2f' % order.executed.price)

            self.bar_executed = len(self)

        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('Order Canceled/Margin/Rejected')

        self.order = None

    # the strategy logic
    def next(self):
        z = self.zscore[0]

        if self.order:
            return
        
        if not self.allow_entry and not self.position:
            if abs(z) < self.entry:
                self.allow_entry = True

        # might buy
        if not self.position and self.allow_entry:
            if z > self.entry and abs(z) < self.stop_loss:
                self.log(f"ENTRY at z = {z:.2f}")
                self.order = self.sell(data=self.datas[0])
                self.order = self.buy(data=self.datas[1])

            elif z < -self.entry and abs(z) < self.stop_loss:
                self.log(f"ENTRY at z = {z:.2f}")
                self.order = self.buy(data=self.datas[0])
                self.order = self.sell(data=self.datas[1])      
        # might sell
        else:
            if abs(z) < self.exit:
                self.log(f"EXIT at z = {z:.2f}")
                self.order = self.close(data=self.datas[0])
                self.order = self.close(data=self.datas[1])

        # stop loss
        if abs(z) > self.stop_loss and self.position:
            self.log(f"STOP‑LOSS at z = {z:.2f}")
            self.order = self.close(data=self.datas[0])
            self.order = self.close(data=self.datas[1])
            self.allow_entry = False
```

```python
# Create a cerebro entity
cerebro = bt.Cerebro()

# Add a strategy
cerebro.addstrategy(StatArb)

data_a = bt.feeds.PandasData(dataname=df_a)
cerebro.adddata(data_a)

data_b = bt.feeds.PandasData(dataname=df_b)
cerebro.adddata(data_b)

# Set our desired cash start
cerebro.broker.setcash(1000)

# commissions
cerebro.broker.setcommission(commission=0.005)

# size of trades
cerebro.addsizer(bt.sizers.PercentSizer, percents=30)

# Print out the starting conditions
print('Starting Portfolio Value: %.2f' % cerebro.broker.getvalue())

# Run over everything
cerebro.run()

# Print out the final result
print('Final Portfolio Value: %.2f' % cerebro.broker.getvalue())

# Plot the result
plt.rcParams['figure.figsize'] = (10, 7)
plt.rcParams['figure.dpi'] = 120
figs = cerebro.plot()
for figlist in figs:
    for f in figlist:
        display(f)
```
